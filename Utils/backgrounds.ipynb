```python
input_folder = "./Letras"
output_folder = "./Fondos"

def gaussian(path, tema, i):
    tema = tema.replace("á", "a")
    tema = tema.replace("í", "i")
    tema = tema.replace("ó", "o")
    tema = tema.replace("ú", "u")
    tema = tema.replace("ñ", "n")
    target = f"{output_folder}/{tema} - {i}.jpg"
    if os.path.exists(target):
        return
    print(target)
    img = cv2.imread(path)
    r = 777
    result = cv2.GaussianBlur(img, (r,r), 0)
    cv2.imwrite(target, result)

def makebg(tema, numero):
    info = numeros[numero]
    path = f'{input_folder}/{info["ruta"]}'
    gaussian(f"{path}/01.jpg", tema, "1")
    gaussian(f"{path}/02.jpg", tema, "2")


if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for tema in temas:
    if type(temas[tema]) == dict:
        for subtema in temas[tema]:
            name = f"{tema} - {subtema}"
            makebg(name, temas[tema][subtema][0])
    else:
        makebg(tema, temas[tema][0])
            
```
